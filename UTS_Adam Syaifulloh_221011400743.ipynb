```python
# Data
company_detail_list = [
    {
        'name': 'Company 1',
        'domain': 'Retail',
        'country': 'United States'
    },
    {
        'name': 'Company 2',
        'domain': 'Technology',
        'country': 'United Kingdom'
    },
    {
        'name': 'Company 3',
        'domain': 'Healthcare',
        'country': 'United States'
    }
]

employee_detail_list = [
    {
        'name': 'EMP-0001',
        'first_name': 'John',
        'last_name': 'Doe',
        'full_name': 'John Doe',
        'company': 'Company 1',
        'nationality': 'Australia'
    },
    {
        'name': 'EMP-0002',
        'first_name': 'Tom',
        'last_name': 'Smith',
        'full_name': 'Tom Smith',
        'company': 'Company 2',
        'nationality': 'United States'
    },
    {
        'name': 'EMP-0003',
        'first_name': 'Andrew',
        'last_name': 'Sebastian',
        'full_name': 'Andrew Sebastian',
        'company': 'Company 3',
        'nationality': 'United States'
    },
    {
        'name': 'EMP-0005',
        'first_name': 'Ying Han',
        'last_name': 'Tan',
        'full_name': 'Ying Han Tan',
        'company': 'Company 1',
        'nationality': 'Australia'
    },
    {
        'name': 'EMP-0015',
        'first_name': 'Kenneth',
        'last_name': 'Ng',
        'full_name': 'Kenneth Ng',
        'company': 'Company 3',
        'nationality': 'United States'
    },
    {
        'name': 'EMP-0018',
        'first_name': 'Rubby',
        'last_name': 'Lee',
        'full_name': 'Rubby Lee',
        'company': 'Company 2',
        'nationality': 'Hong Kong'
    },
    {
        'name': 'EMP-0017',
        'first_name': 'Robert',
        'last_name': 'White',
        'full_name': 'Robert White',
        'company': 'Company 1',
        'nationality': 'United Kingdom'
    }
]

# Task 1: Mendapatkan daftar perusahaan dan mengurutkannya secara terbalik.
def get_sorted_companies():
    sorted_companies = sorted(company_detail_list, key=lambda x: x['name'], reverse=True)
    return [{'name': company['name']} for company in sorted_companies]

# Task 2: Mencetak semua nilai Domain dari setiap perusahaan.
def print_domains():
    for company in company_detail_list:
        print(f"{company['name']}: {company['domain']} ({company['country']})")

# Task 3: Daftar karyawan berdasarkan domain perusahaan.
def list_employees_by_domain():
    domain_employees = {}
    company_domain_map = {company['name']: company['domain'] for company in company_detail_list}
    for employee in employee_detail_list:
        domain = company_domain_map.get(employee['company'])
        if domain:
            if domain not in domain_employees:
                domain_employees[domain] = []
            domain_employees[domain].append(employee['full_name'])
    return domain_employees

# Task 4: Fungsi yang mengembalikan karyawan dengan negara perusahaan mereka.
def employees_with_company_country():
    employees_country = []
    company_country_map = {company['name']: company['country'] for company in company_detail_list}
    for employee in employee_detail_list:
        country = company_country_map.get(employee['company'])
        if country:
            employees_country.append({
                'full_name': employee['full_name'],
                'company': employee['company'],
                'country': country
            })
    return employees_country

# Task 5: Fungsi yang mengembalikan perusahaan dengan jumlah karyawan berdasarkan kewarganegaraan.
from collections import defaultdict

def company_nationality_count():
    company_nationality = defaultdict(lambda: defaultdict(int))
    for employee in employee_detail_list:
        company_nationality[employee['company']][employee['nationality']] += 1
    result = [{'company': company, 'employee_nationality': dict(nat_count)} for company, nat_count in company_nationality.items()]
    return result

# Menampilkan hasil dari setiap tugas
print("Task 1 Result:\n", get_sorted_companies())
print("\nTask 2 Result:")
print_domains()
print("\nTask 3 Result:\n", list_employees_by_domain())
print("\nTask 4 Result:\n", employees_with_company_country())
print("\nTask 5 Result:\n", company_nationality_count())
```
